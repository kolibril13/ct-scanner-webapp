```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
with open("tooth_discrete.json", 'r') as f:
  data = json.load(f)

a = np.zeros((309,309))
for i in data.values():
    a += np.array(i)
plt.imshow(np.clip((a)/40, 0,1), cmap="gray")
plt.axis("off")
```

```python
with open("tooth_continuous.json", 'r') as f:
  data = json.load(f)

a = np.zeros((309,309))
for i in data.values():
    a += np.array(i)
plt.imshow(np.clip((a)/20, 0,1), cmap="gray")
plt.axis("off")
```
