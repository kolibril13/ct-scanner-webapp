```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
with open("../legs_discrete.json", 'r') as f:
  data = json.load(f)
data = data["imgs"]
a = np.zeros((360,360))
for i in data.values():
    a += np.array(i)
plt.imshow(a, cmap="gray")
#plt.imshow(np.clip((a)/40, 0,1), cmap="gray")
plt.axis("off")
```
