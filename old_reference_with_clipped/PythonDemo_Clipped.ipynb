# Discrete

```python
path = Path.cwd() / "imgs_discrete_clipped"

%config InlineBackend.print_figure_kwargs = {'pad_inches':0} # NOT

imgstack = []

for p in sorted(path.iterdir()):
    imgstack.append(plt.imread(p))

imshape = imgstack[0].shape
```

```python
%%capture_img --path "demo_imgs_clipped/discr_imgA.png"
imgA = imgstack[0] 
plt.imshow(imgA)
plt.axis("off")
```

```python
%%capture_img --path "demo_imgs_clipped/discr_imgB.png"

imgstackB = [imgstack[0], imgstack[10]]

imgB = np.zeros(imshape)
for im in imgstackB:
    imgB += im / len(imgstackB)

plt.imshow(imgB)
plt.axis("off")
```

```python
%%capture_img --path "demo_imgs_clipped/discr_imgC.png"

imgstackC = [imgstack[0], imgstack[1],imgstack[2],imgstack[3],imgstack[4],imgstack[5],imgstack[6],imgstack[7],imgstack[8]]

imgC = np.zeros(imshape)
for im in imgstackC:
    imgC += im / len(imgstackC)

plt.imshow(imgC)
plt.axis("off")
```

```python
%%capture_img --path "demo_imgs_clipped/discr_imgD.png"

imgstackD = imgstack

imgD = np.zeros(imshape)
for im in imgstackD:
    imgD += im / len(imgstackD)

plt.imshow(imgD, cmap="gray")
plt.axis("off")
```
